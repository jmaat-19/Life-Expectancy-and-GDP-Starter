```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

life_exp_gdp = pd.read_csv('/Users/jonathanmaat/data_science/Life-Expectancy-and-GDP-Starter/all_data.csv')
print(life_exp_gdp.head(19))
```

```python
print(life_exp_gdp.info())
#there are 96 rows of data and no missing values
```

```python
life_exp_gdp.columns = ['country', 'year', 'average_age', 'gdp']
print(life_exp_gdp.head())
```

```python
print(life_exp_gdp['country'].unique())

# there are 6 different countries in this data set
```

```python
print(96/6)
#want to check that each country has the same data point from year: 2000-2015

for i in life_exp_gdp['country'].unique():
    life_exp_gdp_country = life_exp_gdp[life_exp_gdp.country == i]
    all_years = life_exp_gdp_country.year.unique()
    years_2000_2015 = np.array([2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015])
    if np.array_equal(all_years, years_2000_2015):
        print(i, ' has all 16 values from 2000-2015')
    else:
        print(i, ' has missing values.')
    
    
```

```python
# let's make two individual graphs for each country in subplots to compare gdp and average_age over the years 2000-2016

# first let's make some plots for Chile 
life_exp_gdp_chile = life_exp_gdp[life_exp_gdp.country == 'Chile']
life_exp_gdp_china = life_exp_gdp[life_exp_gdp.country == 'China']
life_exp_gdp_germ = life_exp_gdp[life_exp_gdp.country == 'Germany']
life_exp_gdp_mex = life_exp_gdp[life_exp_gdp.country == 'Mexico']
life_exp_gdp_usa = life_exp_gdp[life_exp_gdp.country == 'United States of America']
life_exp_gdp_zimb = life_exp_gdp[life_exp_gdp.country == 'Zimbabwe']

plt.plot(life_exp_gdp_chile.year, life_exp_gdp_chile.average_age)
plt.plot(life_exp_gdp_china.year, life_exp_gdp_china.average_age)
plt.plot(life_exp_gdp_germ.year, life_exp_gdp_germ.average_age)
plt.plot(life_exp_gdp_mex.year, life_exp_gdp_mex.average_age)
plt.plot(life_exp_gdp_usa.year, life_exp_gdp_usa.average_age)
plt.plot(life_exp_gdp_zimb.year, life_exp_gdp_zimb.average_age)

plt.xlabel('Year')
plt.ylabel('Age')
plt.title('Year vs Age')
plt.legend(['Chile','China','Germany','Mexico','United States of America','Zimbabwe'])
plt.axis([2000, 2015, 44, 82])


plt.show()
plt.clf()


```

```python
#i think it would be good to normalize all axis taking the maxes and min and making that the scale for all the graphs
print(life_exp_gdp.head())

age_col = life_exp_gdp["average_age"]
max_value_age = age_col.max()
min_value_age = age_col.min()
print(max_value_age)
print(min_value_age)

#i think it would also look good to normalize the max and min for gdp
gdp_col = life_exp_gdp['gdp']
max_value_gdp = gdp_col.max()
min_value_gdp = gdp_col.min()
print(max_value_gdp)
print(min_value_gdp)

```

```python
years_list = []
for j in range(16):
    elm = 2000+j
    years_list.append(elm)


plt.figure(figsize=(10,7))
ax = plt.subplot()
ax.set_xticks(years_list)

# gdp graphs for 6 countries
country_list = ['Chile','China','Germany','Mexico','United States of America','Zimbabwe']


for i in country_list:
    life_exp_gdp_country = life_exp_gdp[life_exp_gdp.country == i]
    plt.plot(life_exp_gdp_country.year, life_exp_gdp_country.gdp)
    

plt.xlabel('Year')
plt.xticks(rotation = 90)
plt.ylabel('Gdp (ten trillions)')
plt.title('Year vs Gdp')
plt.legend(['Chile','China','Germany','Mexico','United States of America','Zimbabwe'])
plt.axis([2000, 2015,4415702800.0,18100000000000.0 ])





plt.show()
plt.clf()
```
